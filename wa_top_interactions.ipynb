# Reporting on the top Welcome Ambassador enquiries

Every six weeks, Welcome Ambassadors have a volunteers meeting led by two Volunteer Coordinators (VCs). The VCs like to share information such as what the most common questions have been, which typically has been about specific shops, attractions or locations. In this case, the VCs also like to share what businesses they tend to direct visitors to.

The information and graphics in this Notebook focus on the above two questions to supplement the map showing where in the city centre they tend to log interactions.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# A CSV file will be exported from an external CRM before each WA volunteer meeting.
df = pd.read_csv("welcomehostinteractions_march_filter.csv") # change the file name based on the month desired
df.head(5)
```

```python
# get top type of enquiries
count_q = df["Enquiry"].value_counts()
count_q
```

```python
count_q_plot = count_q.plot.pie(figsize=(15, 15))
plt.pie(count_q,
        autopct='%1.1f%%',
       pctdistance=0.8)
plt.show()
```

```python
# get more detail on top type of enquiry, usually it is "Where is business location?"
top_q = df[df["Enquiry"]=="Specific Shop/Attraction"]
top_q.tail(5)
```

```python
top_biz = top_q["Businesses"].value_counts()
top_biz.head(10)
```

```python
count_biz_plot = top_biz.head(10).plot.pie(figsize=(10, 10))
plt.pie(top_biz.head(10),
        autopct='%1.1f%%')
plt.show()
```
